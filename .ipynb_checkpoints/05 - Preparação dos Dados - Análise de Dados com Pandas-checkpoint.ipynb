# Preparação dos Dados

## Concatenação de Datasets

```python
import pandas as pd
```

```python
df1 = pd.read_csv('concat_1.csv')
df2 = pd.read_csv('concat_2.csv')
df3 = pd.read_csv('concat_3.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#concatenando os 3 datasets - ele empilha os 3 datasets na sequência

concat = pd.concat([df1, df2, df3])

concat
```

```python
## ignorando o índice

concatenar = pd.concat([df1, df2, df3], ignore_index=True)

concatenar
```

## Adicionando Colunas

```python
## por default, o concat mostra axis = 0 o que faz com que os dados sejam concatenados nas linhas, 
## se quisermos por colunas, podemos inserir axis=1

concat_coluna = pd.concat([df1, df2, df3], axis=1)

concat_coluna
```

```python
## filtrando por coluna

concat_coluna['A']
```

```python
## evitando duplicação de nomes nas colunas

concat_coluna = pd.concat([df1, df2, df3], axis=1, ignore_index=True)

concat_coluna
```

## Combinando Diferentes Conjuntos de Dados

```python
person = pd.read_csv('survey_person.csv')
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
## fazendo um subset de dados para evitar duplicação

visited_subset = visited.loc[[0,2,6]]
```

```python
## fazendo um merge - left join

o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')

o2o_merge
```
