# Plotagem com Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Histogramas

```python
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
plt.show()
```

```python
# alterando alguns parâmetros - deixando apenas o histograma

hist, ax = plt.subplots()

ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Valor Total da Conta')
ax.set_ylabel('Frequência')
plt.show()
```

## Plotagem de Densidade com Kernel

```python
den, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Valor Total da Conta')
ax.set_ylabel('Probabilidade Unitária')
plt.show()
```

## Rug Plot - Assemelha as Franjas de um Tapete

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Valor Total da Conta')
plt.show()
```

## Plotagem de Contadores

```python
count, ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Contagem dos Dias')
ax.set_xlabel('Dia da Semana')
ax.set_ylabel('Frequência')
plt.show()
```

# Dados Bivariados - Seaborn

## Gráfico de Dispersão

```python
# caso queira tirar a linha de regressão - inserir fit_reg = False

scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data = tips)
ax.set_title('Dispersão do Valor Total da Conta e Gorjeta')
ax.set_xlabel('Valor Total da Conta')
ax.set_ylabel('Gorjeta')
plt.show()
```

## Plotagem Hexbin

```python
hexbin = sns.jointplot(x= 'total_bill', y='tip', data =tips, kind='hex')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

## Plotagem de Densidade 2D

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                data2=tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bil and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()

kde_joint = sns.jointplot(x = 'total_bill', y ='tip', data = tips, kind = 'kde')
```

## Boxplot

```python
box, ax = plt.subplots()
ax = sns.boxplot(x='time', y ='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('time of day')
ax.set_ylabel('total bill')
plt.show()
```
