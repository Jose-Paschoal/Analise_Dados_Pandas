# Plotagem com Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Histogramas

```python
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
plt.show()
```

```python
# alterando alguns parâmetros - deixando apenas o histograma

hist, ax = plt.subplots()

ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Valor Total da Conta')
ax.set_ylabel('Frequência')
plt.show()
```

## Plotagem de Densidade com Kernel

```python
den, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_xlabel('Valor Total da Conta')
ax.set_ylabel('Probabilidade Unitária')
plt.show()
```

## Rug Plot - Assemelha as Franjas de um Tapete

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Valor Total da Conta')
plt.show()
```

## Plotagem de Contadores

```python
count, ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Contagem dos Dias')
ax.set_xlabel('Dia da Semana')
ax.set_ylabel('Frequência')
plt.show()
```
